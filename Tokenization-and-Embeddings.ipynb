```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
sentences=['My favorite food is ice cream','do you like ice cream too?']
tokenizer=Tokenizer(num_words=10) # optional to set the maximum of 
tokenizer.fit_on_texts(sentences)
```

```python
word_index=tokenizer.word_index
print(word_index)
```

```python
sequences=tokenizer.texts_to_sequences(sentences)
sequences
```

```python
# in case of having a tokens not in the word_index which is already passed so in this case we assing the OOV (OutOfVocab.) to be a unique number
tokenizer=Tokenizer(num_words=20,oov_token='<OOV>')
tokenizer.fit_on_texts(sentences)
```

```python
tokenizer.word_index
```

```python
new_sentence=['Your favorite food is strawberries and cream']
sequence_new=tokenizer.texts_to_sequences(new_sentence)
print(sequence_new) # So in this code words ['Your', 'strawberries', 'and'] not knowen (OOV)
```

```python
sentences = [
'My favorite food is ice cream', 'My lovely girl favorite foode is cheese cake']
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
sequences = tokenizer.texts_to_sequences (sentences)

```
